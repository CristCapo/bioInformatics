```python
import pandas as pd
import re
import time
import numpy as np
```

```python
# Dataset Section
datasetFilename = 'data/Dati HF-LF_3_6_12 mesi.xlsx'
xl = pd.ExcelFile(datasetFilename)
dataFrame = xl.parse('Sheet1')
```

```python
def getGenes(miRNAName):
    geneSeries = dataFrame[dataFrame['miRNA'] == miRNAName]['Target Gene']
    return geneSeries.tolist()
```

```python
def removeLastCharacters(string, charToRemove):
    return string[:-charToRemove]
```

```python
removeLastCharacters('ciaos', 3)
```

```python
def findLastNumbers(columnName):

    result = re.search(r"-[^a-z][0-9]*$", columnName, flags=re.IGNORECASE)

    try:
        return result.group(0)
    except AttributeError:
        return None
```

```python
findLastNumbers('mmu-let-7a-000377')
```

```python
def removeMiRNAIdentifier(columnName):
    result = re.search(r"(.*)(-[^a-z][0-9]*$)", columnName, flags=re.IGNORECASE)
    
    try:
        return result.group(1)
    except AttributeError:
        return None
```

```python
removeMiRNAIdentifier('mmu-let-7a-000377')
```

```python
def getMiRNAList():
    result = []
    resulTwo = []
    miRNAList = dataFrame['Detector']
    start = time.time()
    for element in miRNAList:
        print(element)
        result.append(removeMiRNAIdentifier(element))
    end = time.time()
    print("FIRST:", end - start)

    start = time.time()
    for element in miRNAList:
        resulTwo.append(element[:-7])
    end = time.time()
    print("SECOND:", end - start)

    return result
```

```python
miRNANames = getMiRNAList()
miRNANames = set(miRNANames)
```

```python
# Local Section
datasetFilename = 'data/localDB.xls'
xl = pd.ExcelFile(datasetFilename)
localDF = xl.parse('miRTarBase')
```

```python
miRNADf = pd.DataFrame({'miRNA': list(miRNANames)})
```

```python
resulting = pd.merge(localDF[['miRNA', 'Target Gene']], miRNADf, how = 'inner', on=['miRNA'])
len(set(resulting['miRNA']))
```

```python
def searchForTarget(miRNAName):
    results = resulting['Target Gene'].where(resulting['miRNA'] == miRNAName).dropna()
    return list(set(results.values.tolist()))

searchForTarget('mmu-miR-876-5p')
```
